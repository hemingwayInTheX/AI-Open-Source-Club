```python
import pandas as pd
import numpy as np

df = pd.DataFrame([
                    ['pink','M',14,'class1'],
                    ['orange','S',8,'class2'],
                    ['skyblue','L',19,'class1']
                ])
df.columns = ['color','size','price','classlabel']
df
```

## 1ï¸âƒ£ ìˆœì„œê°€ ìˆëŠ” íŠ¹ì„± ë§¤í•‘

> * sizeíŠ¹ì„±ì„ ìˆœì„œëŒ€ë¡œ ì •ìˆ˜í˜•ìœ¼ë¡œ ì¸ì½”ë”©í•˜ëŠ” í•¨ìˆ˜ëŠ” ì—†ìŒ

> => **ì¸ì½”ë”© í•¨ìˆ˜ ì§ì ‘ ì •ì˜ í•„ìš”**

```python
encoding_1 = {
            'S':1,
            'M':2,
            'L':3,
            }

df['size']=df['size'].map(encoding_1)
df
```

## 2ï¸âƒ£ í´ë˜ìŠ¤ ë ˆì´ë¸” ì¸ì½”ë”©

> * ë¨¸ì‹ ëŸ¬ë‹ ì•Œê³ ë¦¬ì¦˜ì€ í´ë˜ìŠ¤ ë ˆì´ë¸”ì´ ì¸ì½”ë”©ë˜ì—ˆì„ ê±°ë¼ê³  ê°€ì •í•¨

> * => ìˆœì„œ íŠ¹ì„± ë§¤í•‘ê³¼ ë™ì¼í•œ ë°©ì‹ ì ìš©

### ğŸ“ŒMETHOD_1) ë§¤í•‘ ë”•ì…”ë„ˆë¦¬ ì‚¬ìš©í•œ ì¸ì½”ë”©

```python
encoding_2 = {label:idx for idx,label in
             enumerate(np.unique(df['classlabel']))}

encoding_2
```

```python
df['classlabel'] = df['classlabel'].map(encoding_2)
df
```

### ğŸ“ŒMETHOD_2)ì‚¬ì´í‚·ëŸ°ì˜ LabelEncoderí´ë˜ìŠ¤ ì‚¬ìš©


```python
from sklearn.preprocessing import LabelEncoder

#ê°ì²´ ìƒì„±
target = LabelEncoder()

#ë©¤ë²„í•¨ìˆ˜ fit_transform ì‚¬ìš©
result = target.fit_transform(df['classlabel'].values)
result
```
